```python
#@title
import pandas as pd
import seaborn as sns
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
#@title
df = pd.read_json('https://api.flipsidecrypto.com/api/v2/queries/cce74077-7180-4079-9216-aa0379a63330/data/latest')
df = df.loc[df['DATE'] >= '2020-12-03'] # Bridge bolume data only present afterwards
```

```python
#@title
data = df.groupby('DATE').sum().reset_index()
fig = make_subplots(specs=[[{"secondary_y": True}]])
fig.add_trace(go.Bar(x=data['DATE'], y=data['VOLUME_USD'],name='Net USD Volume'))
fig.add_trace(go.Scatter(x=data['DATE'], y=data['TX_COUNT'],name='Transaction Count'),secondary_y=True)
fig.update_layout(
    title="Daily USD Volume and Transaction Count on The Terra ETH Bridge",
    xaxis_title="Date",
    font=dict(
        family="Courier New, monospace",
        size=18,
        color="RebeccaPurple"
    )
    
)
anchor_snapshot = '2021-01-15'
anchor_release = '2021-03-17'
#mirror_snapshot = '2020-11-23'
mirror_ad_start = '2020-12-03'
pylon_snapshot = '2021-06-26'
#pylon_release = 
line_length = 30000000
# Add annotations
fig.add_shape(type="line",
    x0=anchor_snapshot, y0=0, x1=anchor_snapshot, y1=line_length,
    line=dict(
        color="MediumPurple",
        width=4,
        dash="dot",
    )
)
fig.add_shape(type="line",
    x0=anchor_release, y0=0, x1=anchor_release, y1=line_length,
    line=dict(
        color="MediumPurple",
        width=4,
        dash="dot",
    )
)
fig.add_shape(type="line",
    x0=pylon_snapshot, y0=0, x1=pylon_snapshot, y1=line_length,
    line=dict(
        color="MediumPurple",
        width=4,
        dash="dot",
    )
)
fig.add_trace(go.Scatter(
    x=[anchor_snapshot,anchor_release,  pylon_snapshot],
    y=[line_length,line_length, line_length],
    text=["Anchor Snapshot",
          "Anchor Release",
          "Pylon Snapshot"],
    mode="text",
    name='Annotations'
))
fig.update_yaxes(title='Net USD Volume')
fig.update_yaxes(title='Transaction Count',secondary_y=True)
fig.show()
```



*   The daily net bridge USD volume seems to be negative more often than not pre May 29. 
*   After the LUNA price crash this behavior seems to change as the daily net USD volume shifts to being positive.
* I think this shows that people have started to believe in the Terra ecosystem and they are bringing in their money to take part in it.
* Anchor release seems to be a huge step for the ecosystem as there is an enormous peak on March 17 for the number of transactions done with the bridge. There were more than 3000 transactions on a single day!
